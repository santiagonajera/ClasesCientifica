```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Crear datos de ejemplo
np.random.seed(42)
dates = pd.date_range(start='2023-01-01', end='2023-12-31', freq='ME')
data = {
    'Fecha': dates,
    'Producto A': np.random.randint(100, 200, len(dates)),
    'Producto B': np.random.randint(150, 250, len(dates)),
    'Producto C': np.random.randint(80, 180, len(dates))
}

# Crear DataFrame
df = pd.DataFrame(data)

# Configurar el gráfico
plt.figure(figsize=(12, 6))

# Crear el gráfico de área
plt.fill_between(df['Fecha'], df['Producto A'], alpha=0.3, label='Producto A')
plt.fill_between(df['Fecha'], df['Producto B'], alpha=0.3, label='Producto B')
plt.fill_between(df['Fecha'], df['Producto C'], alpha=0.3, label='Producto C')

# Personalizar
plt.title('Ventas por Producto (2023)')
plt.xlabel('Fecha')
plt.ylabel('Unidades Vendidas')
plt.xticks(rotation=45)
plt.legend()
plt.grid(True, alpha=0.3)

# Ajustar layout
plt.tight_layout()

# Mostrar el gráfico
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Crear datos de ejemplo
np.random.seed(42)
dates = pd.date_range(start='2023-01-01', end='2023-12-31', freq='ME')
data = {
    'Fecha': dates,
    'Producto A': np.random.randint(100, 200, len(dates)),
    'Producto B': np.random.randint(150, 250, len(dates)),
    'Producto C': np.random.randint(80, 180, len(dates))
}

# Crear DataFrame
df = pd.DataFrame(data)

# Configurar el gráfico
plt.figure(figsize=(12, 6))

# Crear el gráfico de área apilada
plt.stackplot(df['Fecha'],
             [df['Producto A'], df['Producto B'], df['Producto C']],
             labels=['Producto A', 'Producto B', 'Producto C'],
             alpha=0.8)

# Personalizar
plt.title('Ventas por Producto (2023)')
plt.xlabel('Fecha')
plt.ylabel('Unidades Vendidas')
plt.xticks(rotation=45)
plt.legend(loc='upper left')
plt.grid(True, alpha=0.3)

# Ajustar layout
plt.tight_layout()

# Mostrar el gráfico
plt.show()
```

```python
#GRAFICO PASTEL
# Importar librerías necesarias
import pandas as pd
import matplotlib.pyplot as plt

# Power BI pasa automáticamente los datos cargados al dataframe 'dataset'
# Asegúrate de que los nombres de las columnas coincidan con las de tu tabla en Power BI
dataset = pd.DataFrame({
    'Categoría': ['A', 'B', 'C', 'D'],
    'Valores': [10, 20, 15, 30]
})

# Crear el gráfico de pastel
plt.figure(figsize=(8, 6))
plt.pie(
    dataset['Valores'],
    labels=dataset['Categoría'],
    autopct='%1.1f%%',  # Mostrar los porcentajes con 1 decimal
    startangle=90,  # Rotar para que comience en la parte superior
    colors=['skyblue', 'lightgreen', 'orange', 'lightcoral']
)
plt.title('Gráfico de Pastel en Power BI', fontsize=16)

# Mostrar el gráfico
plt.show()
```

```python
#PASTEL SEABORN
# Importar librerías necesarias
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Power BI pasa automáticamente los datos cargados al dataframe 'dataset'
# Asegúrate de que los nombres de las columnas coincidan con las de tu tabla en Power BI
dataset = pd.DataFrame({
    'Categoría': ['A', 'B', 'C', 'D'],
    'Valores': [10, 20, 15, 30]
})

# Generar una paleta de colores con Seaborn
palette = sns.color_palette('pastel', len(dataset))

# Crear el gráfico de pastel
plt.figure(figsize=(8, 6))
plt.pie(
    dataset['Valores'],
    labels=dataset['Categoría'],
    autopct='%1.1f%%',  # Mostrar los porcentajes con 1 decimal
    startangle=90,  # Rotar para que comience en la parte superior
    colors=palette  # Usar la paleta de colores de Seaborn
)
plt.title('Gráfico de Pastel con Seaborn en Power BI', fontsize=16)

# Mostrar el gráfico
plt.show()
```

```python
import plotly.graph_objects as go

# Configuración de datos
value = 75  # Valor actual
max_value = 100  # Valor máximo
title = "INDICADOR PORCENTAJE CUMPLIMIENTO"
ranges = [
    {"range": [0, 40], "color": "red"},      # Rango bajo
    {"range": [40, 70], "color": "yellow"}, # Rango medio
    {"range": [70, 100], "color": "green"}  # Rango alto
]

# Crear el gráfico de indicador
fig = go.Figure()

# Agregar rangos coloreados al gráfico
for r in ranges:
    fig.add_trace(go.Indicator(
        mode="gauge+number+delta",
        value=value,
        title={"text": title},
        gauge={
            "axis": {"range": [0, max_value], "tickwidth": 1, "tickcolor": "black"},
            "bar": {"color": "gray"},
            "bgcolor": "white",
            "steps": [{"range": r["range"], "color": r["color"]} for r in ranges],
            "threshold": {
                "line": {"color": "black", "width": 4},
                "thickness": 0.65,
                "value": value
            }
        }
    ))

# Personalizar el diseño del gráfico
fig.update_layout(
    font={"family": "Arial", "size": 18},
    margin={"l": 20, "r": 20, "t": 40, "b": 20}
)

# Mostrar el gráfico
fig.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

def stylized_gauge_chart(value, max_value=100, title="Indicador Estilizado"):
    """
    Crear un gráfico de indicadores estilizado utilizando Seaborn y Matplotlib.

    :param value: Valor actual (posición de la aguja)
    :param max_value: Valor máximo de la escala
    :param title: Título del gráfico
    """
    # Configurar los rangos y colores
    ranges = [
        {"start": 0, "end": 40, "color": sns.color_palette("coolwarm", 7)[0]},  # Rojo
        {"start": 40, "end": 70, "color": sns.color_palette("coolwarm", 7)[3]}, # Amarillo
        {"start": 70, "end": 100, "color": sns.color_palette("coolwarm", 7)[5]} # Verde
    ]

    # Configuración del gráfico
    fig, ax = plt.subplots(figsize=(8, 5), subplot_kw={'projection': 'polar'})
    ax.set_theta_zero_location('N')
    ax.set_theta_direction(-1)

    # Dibujar los rangos coloreados
    for r in ranges:
        start_angle = (1 - r["start"] / max_value) * np.pi
        end_angle = (1 - r["end"] / max_value) * np.pi
        ax.barh(0, end_angle - start_angle, left=start_angle, height=0.3, color=r["color"], edgecolor='none')

    # Dibujar la aguja
    needle_angle = (1 - value / max_value) * np.pi
    ax.plot([needle_angle, needle_angle], [0, 0.3], color='black', linewidth=3, label=f"Valor: {value}")

    # Escala y estética
    ax.set_ylim(0, 0.3)
    ax.set_yticks([])
    ax.set_xticks(np.linspace(0, np.pi, 11))
    ax.set_xticklabels([f"{int(x)}" for x in np.linspace(0, max_value, 11)], fontsize=10)
    ax.set_title(title, fontsize=14, pad=20)

    # Leyenda de rangos
    legend_elements = [
        plt.Line2D([0], [0], color=ranges[0]["color"], lw=4, label='Bajo'),
        plt.Line2D([0], [0], color=ranges[1]["color"], lw=4, label='Medio'),
        plt.Line2D([0], [0], color=ranges[2]["color"], lw=4, label='Alto')
    ]
    ax.legend(handles=legend_elements, loc='lower center', bbox_to_anchor=(0.5, -0.2), ncol=3, fontsize=10)

    # Mostrar el gráfico
    plt.tight_layout()
    plt.show()

# Ejemplo de uso
stylized_gauge_chart(value=65, max_value=100, title="Rendimiento KPI")
```
