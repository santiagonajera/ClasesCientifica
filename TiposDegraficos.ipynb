```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Crear datos de ejemplo
np.random.seed(42)
dates = pd.date_range(start='2023-01-01', end='2023-12-31', freq='ME')
data = {
    'Fecha': dates,
    'Producto A': np.random.randint(100, 200, len(dates)),
    'Producto B': np.random.randint(150, 250, len(dates)),
    'Producto C': np.random.randint(80, 180, len(dates))
}

# Crear DataFrame
df = pd.DataFrame(data)

# Configurar el grfico
plt.figure(figsize=(12, 6))

# Crear el grfico de rea
plt.fill_between(df['Fecha'], df['Producto A'], alpha=0.3, label='Producto A')
plt.fill_between(df['Fecha'], df['Producto B'], alpha=0.3, label='Producto B')
plt.fill_between(df['Fecha'], df['Producto C'], alpha=0.3, label='Producto C')

# Personalizar
plt.title('Ventas por Producto (2023)')
plt.xlabel('Fecha')
plt.ylabel('Unidades Vendidas')
plt.xticks(rotation=45)
plt.legend()
plt.grid(True, alpha=0.3)

# Ajustar layout
plt.tight_layout()

# Mostrar el grfico
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Crear datos de ejemplo
np.random.seed(42)
dates = pd.date_range(start='2023-01-01', end='2023-12-31', freq='ME')
data = {
    'Fecha': dates,
    'Producto A': np.random.randint(100, 200, len(dates)),
    'Producto B': np.random.randint(150, 250, len(dates)),
    'Producto C': np.random.randint(80, 180, len(dates))
}

# Crear DataFrame
df = pd.DataFrame(data)

# Configurar el grfico
plt.figure(figsize=(12, 6))

# Crear el grfico de rea apilada
plt.stackplot(df['Fecha'],
             [df['Producto A'], df['Producto B'], df['Producto C']],
             labels=['Producto A', 'Producto B', 'Producto C'],
             alpha=0.8)

# Personalizar
plt.title('Ventas por Producto (2023)')
plt.xlabel('Fecha')
plt.ylabel('Unidades Vendidas')
plt.xticks(rotation=45)
plt.legend(loc='upper left')
plt.grid(True, alpha=0.3)

# Ajustar layout
plt.tight_layout()

# Mostrar el grfico
plt.show()
```

```python
#GRAFICO PASTEL
# Importar libreras necesarias
import pandas as pd
import matplotlib.pyplot as plt

# Power BI pasa automticamente los datos cargados al dataframe 'dataset'
# Asegrate de que los nombres de las columnas coincidan con las de tu tabla en Power BI
dataset = pd.DataFrame({
    'Categora': ['A', 'B', 'C', 'D'],
    'Valores': [10, 20, 15, 30]
})

# Crear el grfico de pastel
plt.figure(figsize=(8, 6))
plt.pie(
    dataset['Valores'],
    labels=dataset['Categora'],
    autopct='%1.1f%%',  # Mostrar los porcentajes con 1 decimal
    startangle=90,  # Rotar para que comience en la parte superior
    colors=['skyblue', 'lightgreen', 'orange', 'lightcoral']
)
plt.title('Grfico de Pastel en Power BI', fontsize=16)

# Mostrar el grfico
plt.show()
```

```python
#PASTEL SEABORN
# Importar libreras necesarias
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Power BI pasa automticamente los datos cargados al dataframe 'dataset'
# Asegrate de que los nombres de las columnas coincidan con las de tu tabla en Power BI
dataset = pd.DataFrame({
    'Categora': ['A', 'B', 'C', 'D'],
    'Valores': [10, 20, 15, 30]
})

# Generar una paleta de colores con Seaborn
palette = sns.color_palette('pastel', len(dataset))

# Crear el grfico de pastel
plt.figure(figsize=(8, 6))
plt.pie(
    dataset['Valores'],
    labels=dataset['Categora'],
    autopct='%1.1f%%',  # Mostrar los porcentajes con 1 decimal
    startangle=90,  # Rotar para que comience en la parte superior
    colors=palette  # Usar la paleta de colores de Seaborn
)
plt.title('Grfico de Pastel con Seaborn en Power BI', fontsize=16)

# Mostrar el grfico
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Configurar datos para el grfico
actual = 800_000_000  # Medida actual
target = 900_000_000  # Medida objetivo
max_value = 1_000_000_000  # Valor mximo del rango
ranges = [400_000_000, 700_000_000, max_value]  # Rangos cualitativos: pobre, satisfactorio, bueno

# Crear el grfico
fig, ax = plt.subplots(figsize=(6, 8))

# Agregar rangos cualitativos con colores pastel (degradado en columnas)
ax.bar(x=0, height=ranges[0], color=sns.color_palette("pastel")[2], edgecolor='none', width=0.5, label='Pobre')
ax.bar(x=0, height=ranges[1] - ranges[0], bottom=ranges[0], color=sns.color_palette("pastel")[1], edgecolor='none', width=0.5, label='Satisfactorio')
ax.bar(x=0, height=ranges[2] - ranges[1], bottom=ranges[1], color=sns.color_palette("pastel")[0], edgecolor='none', width=0.5, label='Bueno')

# Agregar la medida actual (barra principal)
ax.bar(x=0, height=actual, color='cornflowerblue', edgecolor='none', width=0.3, label='Actual')

# Agregar la medida objetivo (lnea de referencia)
ax.axhline(y=target, color='black', linestyle='-', linewidth=2, label='Objetivo')

# Personalizar el grfico
ax.set_xlim(-0.5, 0.5)
ax.set_ylim(0, max_value)
ax.set_xticks([])
ax.set_yticks([0, 200_000_000, 400_000_000, 600_000_000, 800_000_000, 1_000_000_000])
ax.set_yticklabels(['0', '200M', '400M', '600M', '800M', '1B'])
ax.set_ylabel('Revenue', fontsize=10)
ax.set_title('Grfico de Rentabilidad Esperada (Columna)', fontsize=14)

# Reubicar la leyenda de forma horizontal debajo del grfico
ax.legend(
    loc='lower center',
    bbox_to_anchor=(0.5, -0.15),  # Centrar la leyenda debajo
    ncol=5,  # Nmero de columnas en la leyenda
    fontsize=10,
    frameon=False  # Sin cuadro alrededor de la leyenda
)

# Ajustar layout
plt.tight_layout()

# Mostrar el grfico
plt.show()
```

```python
import plotly.graph_objects as go

# Configuracin de datos
value = 75  # Valor actual
max_value = 100  # Valor mximo
title = "INDICADOR PORCENTAJE CUMPLIMIENTO"
ranges = [
    {"range": [0, 40], "color": "red"},      # Rango bajo
    {"range": [40, 70], "color": "yellow"}, # Rango medio
    {"range": [70, 100], "color": "green"}  # Rango alto
]

# Crear el grfico de indicador
fig = go.Figure()

# Agregar rangos coloreados al grfico
for r in ranges:
    fig.add_trace(go.Indicator(
        mode="gauge+number+delta",
        value=value,
        title={"text": title},
        gauge={
            "axis": {"range": [0, max_value], "tickwidth": 1, "tickcolor": "black"},
            "bar": {"color": "gray"},
            "bgcolor": "white",
            "steps": [{"range": r["range"], "color": r["color"]} for r in ranges],
            "threshold": {
                "line": {"color": "black", "width": 4},
                "thickness": 0.65,
                "value": value
            }
        }
    ))

# Personalizar el diseo del grfico
fig.update_layout(
    font={"family": "Arial", "size": 18},
    margin={"l": 20, "r": 20, "t": 40, "b": 20}
)

# Mostrar el grfico
fig.show()
```
