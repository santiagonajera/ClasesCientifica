```python
#Que es un histograma
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Generar datos de ejemplo
np.random.seed(42)
# Simulando alturas en cm
datos = np.random.normal(loc=170, scale=10, size=1000)

# Método 1: Matplotlib básico
plt.figure(figsize=(10, 4))
plt.subplot(121)
plt.hist(datos, bins=30, color='skyblue', edgecolor='black')
plt.title('Histograma con Matplotlib')
plt.xlabel('Valores')
plt.ylabel('Frecuencia')

# Método 2: Seaborn (más estilizado)
plt.subplot(122)
sns.histplot(datos, bins=30, kde=True)
plt.title('Histograma con Seaborn + KDE')
plt.xlabel('Valores')
plt.ylabel('Conteo')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Configurar el tamaño de la figura
plt.figure(figsize=(15, 6))

# Generar datos
np.random.seed(42)
# Datos simétricos (distribución normal)
datos_simetricos = np.random.normal(loc=0, scale=1, size=1000)
# Datos asimétricos (distribución log-normal)
datos_asimetricos = np.random.lognormal(mean=0, sigma=0.5, size=1000)

# Crear subplots
plt.subplot(121)
# Histograma datos simétricos
plt.hist(datos_simetricos, bins=30, color='skyblue', alpha=0.7, density=True)
plt.title(f'Distribución Simétrica\nAsimetría: {stats.skew(datos_simetricos):.2f}')
plt.axvline(np.mean(datos_simetricos), color='red', linestyle='--', label='Media')
plt.axvline(np.median(datos_simetricos), color='green', linestyle='--', label='Mediana')
plt.legend()

plt.subplot(122)
# Histograma datos asimétricos
plt.hist(datos_asimetricos, bins=30, color='lightgreen', alpha=0.7, density=True)
plt.title(f'Distribución Asimétrica\nAsimetría: {stats.skew(datos_asimetricos):.2f}')
plt.axvline(np.mean(datos_asimetricos), color='red', linestyle='--', label='Media')
plt.axvline(np.median(datos_asimetricos), color='green', linestyle='--', label='Mediana')
plt.legend()

plt.tight_layout()
plt.show()

# Análisis estadístico
def analizar_simetria(datos, nombre):
    print(f"\nAnálisis de {nombre}:")
    print(f"Media: {np.mean(datos):.2f}")
    print(f"Mediana: {np.median(datos):.2f}")
    print(f"Asimetría: {stats.skew(datos):.2f}")
    print(f"Diferencia Media-Mediana: {np.mean(datos) - np.median(datos):.2f}")

    # Interpretación de la asimetría
    skewness = abs(stats.skew(datos))
    if skewness < 0.5:
        print("Interpretación: Aproximadamente simétrica")
    elif skewness < 1:
        print("Interpretación: Moderadamente asimétrica")
    else:
        print("Interpretación: Muy asimétrica")

# Analizar ambas distribuciones
analizar_simetria(datos_simetricos, "Distribución Simétrica")
analizar_simetria(datos_asimetricos, "Distribución Asimétrica")
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# Configurar el tamaño de la figura y la semilla aleatoria
np.random.seed(42)

# Crear diferentes distribuciones para ejemplos
def create_distributions(size=1000):
    # Distribución normal (simétrica)
    normal_dist = np.random.normal(loc=0, scale=1, size=size)

    # Distribución asimétrica (log-normal)
    skewed_dist = np.random.lognormal(mean=0, sigma=0.5, size=size)

    # Distribución bimodal
    bimodal_dist = np.concatenate([
        np.random.normal(-2, 0.5, size=size//2),
        np.random.normal(2, 0.5, size=size//2)
    ])

    # Distribución uniforme
    uniform_dist = np.random.uniform(-3, 3, size=size)

    return normal_dist, skewed_dist, bimodal_dist, uniform_dist

# Crear las distribuciones
normal_dist, skewed_dist, bimodal_dist, uniform_dist = create_distributions()

# Crear la figura con subplots
fig, axes = plt.subplots(2, 2, figsize=(15, 12))


# 1. Distribución Normal (Simétrica)
axes[0,0].hist(normal_dist, bins=30, color='skyblue', density=True, alpha=0.7)
sns.kdeplot(normal_dist, ax=axes[0,0], color='darkblue')
axes[0,0].set_title('Distribución Normal (Simétrica)')
axes[0,0].text(-2.5, 0.3, f'Asimetría: {stats.skew(normal_dist):.2f}\n' +
                        f'Curtosis: {stats.kurtosis(normal_dist):.2f}')

# 2. Distribución Asimétrica
axes[0,1].hist(skewed_dist, bins=30, color='lightgreen', density=True, alpha=0.7)
sns.kdeplot(skewed_dist, ax=axes[0,1], color='darkgreen')
axes[0,1].set_title('Distribución Asimétrica (Log-normal)')
axes[0,1].text(2, 0.8, f'Asimetría: {stats.skew(skewed_dist):.2f}\n' +
                        f'Curtosis: {stats.kurtosis(skewed_dist):.2f}')

# 3. Distribución Bimodal
axes[1,0].hist(bimodal_dist, bins=30, color='salmon', density=True, alpha=0.7)
sns.kdeplot(bimodal_dist, ax=axes[1,0], color='red')
axes[1,0].set_title('Distribución Bimodal')
axes[1,0].text(-4, 0.3, f'Asimetría: {stats.skew(bimodal_dist):.2f}\n' +
                      f'Curtosis: {stats.kurtosis(bimodal_dist):.2f}')

# 4. Demostración de diferentes números de bins
bins_options = [5, 50]  # Pocos y muchos bins
colors = ['purple', 'lightgray']
for bins, color in zip(bins_options, colors):
    axes[1,1].hist(normal_dist, bins=bins, alpha=0.5, color=color,
                   label=f'{bins} bins', density=True)
axes[1,1].set_title('Efecto del Número de Bins')
axes[1,1].legend()

# Ajustar el diseño
plt.tight_layout()

# Añadir un texto explicativo general
fig.text(0.02, 0.02, """
Interpretación:
• Distribución Normal: Simétrica alrededor de la media
• Distribución Asimétrica: Cola más larga hacia la derecha
• Distribución Bimodal: Dos picos distintos
• Número de Bins: Afecta la granularidad de la visualización
""", fontsize=10, bbox=dict(facecolor='white', alpha=0.8))

# Mostrar la figura
plt.show()

# Función para analizar una distribución
def analizar_distribucion(datos, nombre="Distribución"):
    """
    Analiza características principales de una distribución
    """
    print(f"\nAnálisis de {nombre}:")
    print(f"Media: {np.mean(datos):.2f}")
    print(f"Mediana: {np.median(datos):.2f}")
    print(f"Desviación estándar: {np.std(datos):.2f}")
    print(f"Asimetría: {stats.skew(datos):.2f}")
    print(f"Curtosis: {stats.kurtosis(datos):.2f}")
    print(f"Test de normalidad (Shapiro-Wilk):")
    stat, p_value = stats.shapiro(datos)
    print(f"p-valor: {p_value:.4f}")
    print("Normal" if p_value > 0.05 else "No normal")

# Analizar cada distribución
distribuciones = {
    "Normal": normal_dist,
    "Asimétrica": skewed_dist,
    "Bimodal": bimodal_dist,
    "Uniforme": uniform_dist
}

for nombre, datos in distribuciones.items():
    analizar_distribucion(datos, nombre)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# Cargar el dataset de tips (propinas en restaurantes)
# Este dataset contiene información económica sobre gastos y propinas
tips = sns.load_dataset('tips')

# Añadir algunas métricas económicas calculadas
tips['porcentaje_propina'] = (tips['tip'] / tips['total_bill']) * 100
tips['gasto_por_persona'] = tips['total_bill'] / tips['size']

# Crear visualización de histogramas
plt.figure(figsize=(15, 10))

# Variables a analizar
variables = ['total_bill', 'gasto_por_persona', 'porcentaje_propina', 'size']
titulos = ['Gasto Total', 'Gasto por Persona', 'Porcentaje de Propina', 'Tamaño del Grupo']
colors = ['lightblue', 'lightgreen', 'salmon', 'purple']

for idx, (var, titulo, color) in enumerate(zip(variables, titulos, colors), 1):
    plt.subplot(2, 2, idx)

    # Histograma con curva de densidad
    plt.hist(tips[var], bins=30, alpha=0.7, color=color, density=True)
    sns.kdeplot(data=tips[var], color='darkblue', linewidth=2)

    # Añadir líneas verticales para media y mediana
    plt.axvline(tips[var].mean(), color='red', linestyle='--', label='Media')
    plt.axvline(tips[var].median(), color='green', linestyle='--', label='Mediana')

    plt.title(f'Distribución de {titulo}\n' +
              f'Asimetría: {stats.skew(tips[var]):.2f}')
    plt.xlabel(titulo)
    plt.ylabel('Densidad')
    plt.grid(True, alpha=0.3)
    if idx == 1:
        plt.legend()

plt.tight_layout()
plt.show()

# Análisis estadístico
def analizar_variable(data, variable, titulo):
    print(f"\nAnálisis de {titulo}:")
    print("-" * 50)
    print(f"Número de observaciones: {len(data[variable])}")
    print(f"Media: ${data[variable].mean():.2f}")
    print(f"Mediana: ${data[variable].median():.2f}")
    print(f"Desviación estándar: ${data[variable].std():.2f}")
    print(f"Coeficiente de variación: {(data[variable].std() / data[variable].mean() * 100):.2f}%")
    print(f"Asimetría: {stats.skew(data[variable]):.2f}")
    print(f"Mínimo: ${data[variable].min():.2f}")
    print(f"Máximo: ${data[variable].max():.2f}")
    print(f"Percentil 25: ${data[variable].quantile(0.25):.2f}")
    print(f"Percentil 75: ${data[variable].quantile(0.75):.2f}")

# Realizar análisis para cada variable
for var, titulo in zip(variables, titulos):
    analizar_variable(tips, var, titulo)

# Análisis de correlación
plt.figure(figsize=(10, 8))
correlation_matrix = tips[variables].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0,
            xticklabels=titulos,
            yticklabels=titulos)
plt.title('Matriz de Correlación entre Variables')
plt.tight_layout()
plt.show()

# Análisis por día de la semana
plt.figure(figsize=(12, 6))
sns.boxplot(x='day', y='gasto_por_persona', data=tips)
plt.title('Distribución del Gasto por Persona según Día de la Semana')
plt.xlabel('Día')
plt.ylabel('Gasto por Persona ($)')
plt.grid(True, alpha=0.3)
plt.show()

# Análisis de tendencias por tiempo de comida
plt.figure(figsize=(10, 6))
sns.violinplot(x='time', y='total_bill', data=tips)
plt.title('Distribución del Gasto Total por Tiempo de Comida')
plt.xlabel('Tiempo de Comida')
plt.ylabel('Gasto Total ($)')
plt.show()

# Análisis de regresión
plt.figure(figsize=(10, 6))
sns.regplot(x='total_bill', y='tip', data=tips, scatter_kws={'alpha':0.5})
plt.title('Relación entre Gasto Total y Propina')
plt.xlabel('Gasto Total ($)')
plt.ylabel('Propina ($)')
plt.grid(True, alpha=0.3)
plt.show()
```
