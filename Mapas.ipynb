```python
import folium
import matplotlib.pyplot as plt
from io import BytesIO
import base64

# Función para crear un gráfico de pastel y convertirlo a una imagen base64
def crear_grafico_pastel(proporcion_hombres, proporcion_mujeres):
    fig, ax = plt.subplots(figsize=(4, 4))
    labels = 'Hombres', 'Mujeres'
    sizes = [proporcion_hombres, proporcion_mujeres]
    colors = ['#ff9999','#66b3ff']
    ax.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
    ax.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

    # Guardar la figura en un buffer
    buf = BytesIO()
    plt.savefig(buf, format='png')
    plt.close(fig)
    buf.seek(0)

    # Convertir la imagen a base64
    img_base64 = base64.b64encode(buf.read()).decode('utf-8')
    return img_base64

# Crear un mapa centrado en una ubicación específica
mapa = folium.Map(location=[-2.1962, -79.8862], zoom_start=5)

# Datos de ejemplo para los marcadores y gráficos de pastel
ciudades = [
    {"nombre": "Lima", "latitud": -12.0464, "longitud": -77.0428, "hombres": 50, "mujeres": 50},
    {"nombre": "Quito", "latitud": -0.1807, "longitud": -78.4678, "hombres": 45, "mujeres": 55},
    {"nombre": "Bogotá", "latitud": 4.7110, "longitud": -74.0721, "hombres": 55, "mujeres": 45}
]

# Añadir marcadores al mapa con gráficos de pastel
for ciudad in ciudades:
    grafico_base64 = crear_grafico_pastel(ciudad["hombres"], ciudad["mujeres"])
    html = f'''
    <div style="text-align:center;">
        <h3>{ciudad["nombre"]}</h3>
        <img src="data:image/png;base64,{grafico_base64}" style="width:200px;height:200px;">
    </div>
    '''
    iframe = folium.IFrame(html, width=250, height=250)
    popup = folium.Popup(iframe, max_width=250)
    folium.Marker(
        location=[ciudad["latitud"], ciudad["longitud"]],
        popup=popup,
        tooltip=ciudad["nombre"]
    ).add_to(mapa)

# Crear una línea que conecte los puntos
coordenadas = [(ciudad["latitud"], ciudad["longitud"]) for ciudad in ciudades]
folium.PolyLine(coordenadas, color="blue", weight=2.5, opacity=1).add_to(mapa)

# Guardar el mapa en un archivo HTML
mapa.save('mapa.html')

# Mostrar el mapa en un Jupyter Notebook (opcional)
mapa
```

```python
import folium
import matplotlib.pyplot as plt
from io import BytesIO
import base64

# Función para crear un gráfico de líneas y convertirlo a una imagen base64
def crear_grafico_lineas(anios, poblacion):
    fig, ax = plt.subplots(figsize=(4, 4))
    ax.plot(anios, poblacion, marker='o', color='b')
    ax.set_title('Crecimiento Poblacional (2018-2024)')
    ax.set_xlabel('Año')
    ax.set_ylabel('Población (millones)')
    ax.grid(True)

    # Guardar la figura en un buffer
    buf = BytesIO()
    plt.savefig(buf, format='png', bbox_inches='tight')
    plt.close(fig)
    buf.seek(0)

    # Convertir la imagen a base64
    img_base64 = base64.b64encode(buf.read()).decode('utf-8')
    return img_base64

# Crear un mapa centrado en una ubicación específica
mapa = folium.Map(location=[-2.1962, -79.8862], zoom_start=5)

# Datos de ejemplo para los marcadores y gráficos de líneas
ciudades = [
    {"nombre": "Lima", "pais": "Perú", "latitud": -12.0464, "longitud": -77.0428,
     "poblacion": [9.1, 9.3, 9.5, 9.7, 9.8, 10.0, 10.2]},
    {"nombre": "Quito", "pais": "Ecuador", "latitud": -0.1807, "longitud": -78.4678,
     "poblacion": [2.7, 2.8, 2.9, 3.0, 3.1, 3.2, 3.3]},
    {"nombre": "Bogotá", "pais": "Colombia", "latitud": 4.7110, "longitud": -74.0721,
     "poblacion": [7.8, 8.0, 8.2, 8.4, 8.6, 8.8, 9.0]},
    {"nombre": "Santiago", "pais": "Chile", "latitud": -33.4489, "longitud": -70.6693,
     "poblacion": [5.6, 5.7, 5.8, 5.9, 6.0, 6.1, 6.2]},
    {"nombre": "Buenos Aires", "pais": "Argentina", "latitud": -34.6037, "longitud": -58.3816,
     "poblacion": [15.2, 15.4, 15.6, 15.8, 16.0, 16.2, 16.4]}
]

anios = [2018, 2019, 2020, 2021, 2022, 2023, 2024]

# Añadir marcadores al mapa con gráficos de líneas
for ciudad in ciudades:
    grafico_base64 = crear_grafico_lineas(anios, ciudad["poblacion"])
    html = f'''
    <div style="text-align:center;">
        <h3>{ciudad["nombre"]}</h3>
        <img src="data:image/png;base64,{grafico_base64}" style="width:300px;height:300px;">
    </div>
    '''
    iframe = folium.IFrame(html, width=350, height=350)
    popup = folium.Popup(iframe, max_width=350)
    folium.Marker(
        location=[ciudad["latitud"], ciudad["longitud"]],
        popup=popup,
        tooltip=f'{ciudad["nombre"]}, {ciudad["pais"]}'
    ).add_to(mapa)

# Guardar el mapa en un archivo HTML
mapa.save('mapa.html')

# Mostrar el mapa en un Jupyter Notebook (opcional)
mapa
```
